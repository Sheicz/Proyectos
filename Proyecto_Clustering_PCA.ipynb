```python
dataset = '/content/drive/MyDrive/dataset_tissue.txt'

df = pd.read_csv(dataset, delimiter=',', index_col=0)
```

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

def apply_pca(dataframe, n_components=2):

    # Estandarizar los datos (importante para PCA)
    scaler = StandardScaler()
    scaled_data = scaler.fit_transform(dataframe)

    # Aplicar PCA
    pca = PCA(n_components=n_components)
    principal_components = pca.fit_transform(scaled_data)

    # Crear un nuevo DataFrame con los componentes principales
    columns = [f'PC{i+1}' for i in range(n_components)]
    result_df = pd.DataFrame(data=principal_components, columns=columns)

    return result_df
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler


min_max_scaler = MinMaxScaler()
normalized_features= min_max_scaler.fit_transform(df)
```

```python
from sklearn.decomposition import PCA

# Reducir la dimensionalidad a 2 dimensiones (puedes ajustar este número según lo que necesites)
pca = PCA(n_components=80)
data_pca = pca.fit_transform(normalized_features)
```

```python
data_pca
```

```python
from sklearn.mixture import GaussianMixture

gmm = GaussianMixture(n_components=8)  # El número de componentes es el número de clusters
gmm_labels = gmm.fit_predict(data_pca)
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=8)
kmeans_labels = kmeans.fit_predict(data_pca)
```

```python
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan_labels = dbscan.fit_predict(data_pca)
```

```python
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score

# Ejemplo para K-Means:
silhouette_kmeans = silhouette_score(data_pca, kmeans_labels)
calinski_kmeans = calinski_harabasz_score(data_pca, kmeans_labels)
davies_kmeans = davies_bouldin_score(data_pca, kmeans_labels)
```

```python
import matplotlib.pyplot as plt

plt.scatter(data_pca[:, 0], data_pca[:, 1], c=kmeans_labels, cmap='rainbow')
plt.title('K-Means Clustering')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

```python
plt.scatter(data_pca[:, 0], data_pca[:, 1], c=gmm_labels, cmap='rainbow')
plt.title('GMM Clustering')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

```python
plt.scatter(data_pca[:, 0], data_pca[:, 1], c=dbscan_labels, cmap='rainbow')
plt.title('DBSCAN Clustering')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```
